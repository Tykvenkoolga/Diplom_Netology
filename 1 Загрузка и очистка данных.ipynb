Заменим пропуски в колонке "Возраст перед отдачей" на разницу между датой рождения и сроком выбытия из приюта

```python
#Переводим даты в формат datetime
all['DateTime_in'] = pd.to_datetime(all['DateTime_in']) 
all['DateTime_out'] = pd.to_datetime(all['DateTime_out'])
all['Date of Birth'] = pd.to_datetime(all['Date of Birth'])
```

Разделим колонки "Sex upon Intake" и "Sex upon Outcome" на колонки, обозначающие пол и статус стерильности

```python
def gender(val):
    try:
        val_list = val.split(' ')
        return val_list [1]
    except:
        return np.nan

def sterill(val):
    try:
        val_list = val.split(' ')
        return val_list [0]
    except:
        return np.nan
```

```python
all['Gender_in'] = all['Sex upon Intake'].apply(gender)
all['Sterill_in'] = all['Sex upon Intake'].apply(sterill)
all['Gender_out'] = all['Sex upon Outcome'].apply(gender)
all['Sterill_out'] = all['Sex upon Outcome'].apply(sterill)
all.head()
```

```python
all['Gender_in'] = all['Gender_in'].fillna('Unknown')
all = all.drop(['Gender_out', 'Sex upon Intake', 'Sex upon Outcome'], axis = 1)
all.head()
```

Рассортируем колонку "Age upon Intake" по категориям

```python
def get_age(row):
    age_string = row['Age upon Intake']
    [age,unit] = age_string.split(" ")
    age = int(age)
    unit = unit.lower()
    if ('day' in unit):
            return 'Less than 1 week'
    elif ('week' in unit):
            return '1-6 weeks'
    elif ('month' in unit):
            if (age)<=6:
                return '1-6 months'
            else:
                return '7-12 months'
    elif ('year' in unit):
            if (age)<=3:
                return '1-3 years'
            elif (age)>3 and (age)<=6:
                return '4-6 years'
            elif (age)>6 and (age)<=9:
                return '7-9 years'
            elif ('year' in unit):
                return '9+ years'
all['group of age_in'] = all.apply(get_age, axis=1)
all.head()
```

Создадим новую колонку и посчитаем, сколько дней животное находилось в приюте перед отбытием из него

# Проблема: в колонке есть отрицательные значения, что с ними делать?

```python
all['days_of_stay'] = (all['DateTime_out'] - all['DateTime_in']).dt.days
all.head()
```

Сгруппируем время, проведенное в приюте по категориям

```python
import re
def days_stay (val):
    val = str(val)
    days = re.findall('\d*',val)[0]
    try:
        days = int(days)
        if days <= 7:
            return "0-7 days"
        elif days <= 21:
            return "1-3 weeks"
        elif days <= 42:
            return "3-6 weeks"
        elif days <= 84:
            return "7-12 weeks"
        elif days <= 168:
            return "12 weeks - 6 months"
        elif days <= 365:
            return "6-12 months"
        elif days <= 730:
            return "1-2 years"
        else:
            return "2+ years"
    except:
        return np.nan
all['group of stay'] = all['days_of_stay'].apply(days_stay)
all.head()
```

Так как в колонках Sterill_in и Sterill_out значения Neutered и Spayed обозначают одно и то же, заменим значения на 0 и 1, 
где 0 - нестерилизованное животное, 1 - стерилизованное

```python
exter_dict_d = ['Sterill_in']
exter_dict = { 'Sterill_in': {'Intact':0, 'Neutered':1, 'Spayed':1}}
all.replace(exter_dict, inplace=True)
all[exter_dict_d] = all[exter_dict_d].fillna(0)
all.head()
```

```python
exter_dict_d = ['Sterill_out']
exter_dict = { 'Sterill_out': {'Intact':0, 'Neutered':1, 'Spayed':1}}
all.replace(exter_dict, inplace=True)
all[exter_dict_d] = all[exter_dict_d].fillna(0)
all.head()
```

Заменим значения в колонке "Name_in" на 0 и 1, где 0 - нет имени, 1 - есть имя

```python
def name(row):
    name_string = row['Name_in']
    if ('Unknown' in name_string):
            return '0'
    else:
            return '1'
all['Name_in'] = all.apply(name, axis=1)
all.head()
```
